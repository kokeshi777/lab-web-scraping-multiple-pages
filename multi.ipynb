```python

time=[]
lat=[]
lon=[]
region=[]
```

```python
for i in range(20): 
    time.append(soup.select("td.tabev6")[i].get_text())
    lat.append(soup.select("td:nth-child(5)")[i].get_text())
    lon.append(soup.select("td:nth-child(7)")[i].get_text())
    region.append(soup.select("td:nth-child(12)")[i].get_text())
    

```

```python
eq_df = pd.DataFrame({"time":time,
                          "lat":lat,
                          "long": lon,
                            "region_name": region})
eq_df.head()
```

```python
eq_df['time'] = list(map(lambda x: x.replace("earthquake",""),eq_df['time']))
eq_df.head()
```

```python
eq_df['time'] = list(map(lambda x: x.replace("min ago",""),eq_df['time']))
eq_df.head()
```

```python
eq_df.time = eq_df.time.astype(str)
```

```python
eq_df.time.str.split("",expand=True,)
```

```python
eq_df[['date','time']] = eq_df.time.str.split(", ",expand=True,)
eq_df.head()
```
